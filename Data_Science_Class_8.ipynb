```python
import pandas as pd
import numpy as np
data=pd.read_csv("netflix_data.csv")

data['cast'].replace(np.nan,"None",inplace=True)

def cast_count(val):
    if val == "none":
        return 0
    else:
        return len(str(val).split(", "))
    
data["num_of_cast"]= data["cast"].apply(cast_count)
data
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='#000',linestyle='solid')
plt.plot(x,np.sin(x-1),color='chartreuse',linestyle='--')
plt.plot(x,np.sin(x-2),color='g',linestyle=':')
plt.plot(x,np.sin(x-3),color='b',linestyle='-.')
plt.plot(x,np.sin(x-4),':r')
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='#000',linestyle='solid')
plt.plot(x,np.sin(x-1),color='chartreuse',linestyle='--')
plt.plot(x,np.sin(x-2),color='g',linestyle=':')
plt.plot(x,np.sin(x-3),color='b',linestyle='-.')
plt.plot(x,np.sin(x-4),':r')

plt.title("Graph of Sin")
plt.xlabel("x")
plt.ylabel('sin(x)')
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='#000',linestyle='solid',label='x')
plt.plot(x,np.sin(x-1),color='chartreuse',linestyle='--',label='x-1')
plt.plot(x,np.sin(x-2),color='g',linestyle=':',label='x-2')
plt.plot(x,np.sin(x-3),color='b',linestyle='-.',label='x-3')
plt.plot(x,np.sin(x-4),':r',label='x-4')

plt.title("Graph of Sin")
plt.xlabel("x")
plt.ylabel('sin(x)')

plt.legend()
plt.show()
```
